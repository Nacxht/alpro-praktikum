# 1. Apa itu Flowchart?

Flowchart adalah diagram untuk menggambarkan suatu urutan langkah-langkah yang diperlukan untuk menjalankan suatu proses secara logis, terstruktur, dan programmatik.

# 2. Apa keuntungan kita membuat Flowchart sebelum menulis kode?

Keuntungannya adalah:
- Dapat menulis kode tanpa perlu pusing memikirkan langkah-langkah, karena langkah-langkahnya sudah dirancang di dalam flowchart.
- Kode lebih efisien, karena kita sudah merancang algoritma di flowchart terlebih dahulu
