# 1. Buatlah fungsi jumlahDigit(n) yang menerima sebuah angka n dan mengembalikan suatu nilai yaitu jumlah dari semua digit (berupa integer) pada angka tersebut. Gunakan loop untuk menjumlahkan setiap digit. contoh jika kita memasukan angka 4523 sebagai parameter maka akan mengembalikan nilai 14
Note: inputan merupakan integer

contoh output:
> image.png

```python
def jumlahDigit(num):
  sum = 0

  while num > 0:
    sum += (num % 10)
    num //= 10

  return sum

print(jumlahDigit(5555))
```

# 2. Buat fungsi angkaTerbalik(n) yang menerima sebuah angka integer n dan yang akan mencetak kebalikan dari angka tersebut, contoh: jika n = 100 maka yang di cetak adalah 001

Note: inputan merupakan integer

contoh output:
> image.png

```python
def angkaTerbalik(num):
  numTemp = num
  reversed = 0
  length = 0

  while num > 0:
    length += 1
    num //= 10

  num = numTemp

  while num > 0:
    reversed = (reversed * 10) + (num % 10)
    num //= 10

  print(f"kebalikan dari {numTemp} merupakan {reversed:0{length}d}")

angkaTerbalik(100)
```

# 3. Buat fungsi cetakSegitiga(n) yang menerima satu parameter n dan mencetak deret angka dari 0 hingga 9 jika angka lebih dari sembilan angka akan kembali menjadi 0 dalam pola segitiga. Gunakan loop untuk membentuk deret tersebut.

contoh output:
> image.png

```python
def cetakSegitiga(length):
  multiplier = 1
  temp = 0
  i = 0

  while i < length:
    if temp > 9:
      temp = 0

    print(" " * (length - i), end="")
    print(str(temp) * multiplier)

    multiplier += 2
    temp += 1
    i += 1

cetakSegitiga(15)
```

# 4. buat fungsi pangkat(a, b) menerima parameter integer dan mengembalikan nilai integer yaitu hasil dari a pangkat b:

contoh output:
>image.png

```python
def pangkat(num, power):
  result = num ** power
  return result

print(pangkat(5, 2))
```
