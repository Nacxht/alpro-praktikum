# 1. Jelaskan cara instalasi python pada perangkat anda.

Berikut adalah penjelasan bagaimana saya menginstall python pada perangkat saya:

Di laptop saya terdapat 2 OS, yaitu fedora & windows.

- Fedora:
> Saat pertama kali menginstlal Fedora, Python sudah secara default terinstall. Tetapi tetap terdapat cara manual untuk menginstall python, berikut caranya:
  1. Buka terminal.
  2. Ketik ```sudo dnf install python3 python-devel```.
  3. Masukkan sandi anda, karena operasi ini membutuhkan hak akses admin.
  4. Jika diberi pertanyaan ingin menginstall atau tidak (y/N), ketik "y".
  5. Setelah itu, kita tinggal menunggu prosesnya selesai.
  6. Jika proses telah selesai, Python sudah terinstal.

- Windows:
> Jika di Windows, saya perlu memasang Python secara manual. Berikut langkah-langkahnya:
  1. Download Python di web resminya ([Klik Disini](https://www.python.org/)).
  2. Setelah file terinstall, klik 2x setup Python yang telah didownload.
  3. Klik "Install Now".
  4. Tunggu hingga proses selesai.
  5. Selamat, Python sudah terinstall!


# 2. Jelaskan tiga jenis algoritma pada pemrograman.

Algoritma pemrograman terdapat 3 jenis, yaitu:
1. Sequence
> Sequence berarti berurutan, yang artinya program dijalankan secara berurutan dari atas ke bawah.
2. Branches/Percabangan
> Percabangan merupakan alur yang dibagi menjadi 2, yang dimana jalan kita ditentukan benar/tidaknya suatu kondisi yang telah ditentukan.
3. Looping/Perulangan
> Perulangan merupakan alur yang diulang sesuai jumlah yang ditentukan. Perulangan terdapat 3 jenis, yaitu `for`, `while`, `do-while`. Masing-masing jenis loop memiliki karakteristiknya tersendiri.

Level dalam bahasa pemrograman memiliki tiga tingkatan, yaitu:
1. High-level:
> Bahasa pemrograman dengan tingkatan high-level sangat mudah dipahami oleh manusia, dikarenakan sintaksnya menggunakan bahasa manusia. Berikut kelebihan dan kekurangannya:
  - Kelebihan:
    - Dapat dengan mudah dipahami oleh manusia.
    - Dapat dengan mudah dipelajari.
  - Kekurangan:
    - Kecepatan lambat.
    - Tidak sepenuhnya terhubung dengan mesin.
2. Mid-level:
> Bahasa pemrograman dengan tingkatan mid-level menurut saya adalah campuran dari high-level dan low-level, karena bahasanya memang menggunakan bahasa mesin yang tidak dapat dipahami oleh manusia. Berikut kelebihan dan kekurangannya:
  - Kelebihan:
    - Performa lebih cepat dibanding dari high-level.
    - Efisien, karena kita masih diberikan kontrol untuk mengatur memori usage.
  - Kekurangan:
    - Bahasa masih bisa dibilang kompleks.
    - Lebih rentan terhadap kesalahan. Contohnya dikarenakan kita diberi kontrol untuk mengatur penggunaan memori, kemungkinan terjadi kebocoran memori (penggunaan memori melonjak tinggi) sangat memungkinkan terjadi.
3. Low-level:
> Bahasa pemrograman dengan tingkatan low-level sangat berbeda dengan kedua tingkatan lainnya. Di tingkat ini sintaks-sintaksnya sangat susah dipahami oleh manusia, bahkan sudah tidak bisa dibaca dengan bahasa manusia. Bahasa di tingkatan ini sudah berupa bahasa mesin.
  - Kelebihan:
    - Performa sangat cepat dikarenakan menggunakan bahasa mesin, yang berarti tidak perlu dicompile atau diinterpretasikan.
    - Langsung terhubung dengan mesin karena bahasanya langsung terhubung dengan mesin.
  - Kekurangan:
    - Sulit dipahami oleh manusia, bahkan tidak dapat dipahami sama sekali.
    - Sudah jarang dipakai, sehingga mungkin susah mencari bantuan komunitas.

# 4. Dalam bahasa pemrograman Python, jelaskan konsep variabel, tipe data, operator, input dan output.

Berikut penjelasan dari variabel, tipe data, operator, input, dan output:

1. Variabel
> Variabel bisa diibaratkan seperti wadah kosong yang dapat diisi dengan suatu nilai. Jadi fungsi dari variabel adalah untuk menyimpan suatu nilai yang dapat diakses dan digunakan dengan cara menggunakan identitas/nama variabel tersebut. Contoh variabel ```nama = "Agus"```, variabel tersebut menyimpan suatu nama seseorang bernama "Agus".
2. Tipe data
> Tipe data bisa dibilang berfungsi untuk memberikan pembeda tiap nilai, seperti `integer` hanya bisa diisi oleh `angka`, yang berarti integer tidak bisa diisi oleh tipe lain seperti kata-kata atau huruf. Beberapa tipe data utama dalam Python:
 - String: tipe data untuk karakter, seperti huruf, dan tulisan. Angka pun dapat dimasukkan ke dalam tipe data String.
 - Integer: tipe data ini digunakan untuk bilangan bulat.
 - Float: tipe data ini digunakan untuk bilangan desimal, cocok digunakan untuk perhitungan yang mengeluarkan hasil desimal (luas lingkaran, luas tabung, dll).
 - Boolean: tipe data ini merupakan tipe data yang hanya memiliki 2 nilai, yaitu `True` & `False`.
3. Operator
> Operator merupakan simbol atau kata kunci untuk melakukan suatu operasi pada suatu nilai. Beberapa jenis operator:
 - Operator aritmatika: digunakan untuk melakukan proses aritmatika.
 - Operator perbandingan: digunakan untuk melakukan proses perbandingan, dan tipe keluarannya adalah boolean. Contoh (1 > 2 ; hasilnya adalah false).
 - Operator logika: digunakan untuk melakukan perbandingan logika dengan aturan gerbang logika (AND, OR, NOT).
4. Input
> Input adalah suatu proses dimana program meminta user untuk melakukan suatu input yang nantinya user input tersebut akan diolah atau diproses oleh program.
5. Output
> Output adalah keluaran dari suatu program, output bisa digunakan untuk memberikan informasi kepada user atau memberikan hasil proses kepada user.
