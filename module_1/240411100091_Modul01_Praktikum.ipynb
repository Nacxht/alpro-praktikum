# 2. Buat program dengan Bahasa Python untuk konversi jam ke menit dan ke detik seperti yang ditunjukkan pada gambar berikut.
image.png

```python
print("Konversi jam ke menit dan ke detik.")

hrs = int(input("Input dalam format jam: "))
mins = hrs * 60
secs = mins * 60

# print(f"jam: {type(hrs)}")
# print(f"menit: {type(mins)}")
# print(f"detik: {type(secs)}")

print(f"{hrs} jam ada {mins} menit")
print(f"{hrs} jam: {mins} menit: {secs} detik")
```

# 3. Buat program yang meminta input suhu dalam Celcius, lalu konversi suhu tersebut ke Fahrenheit dan Kelvin, Seperti gambar berikut.

image.png

```python
celcius = float(input("Masukkan suhu dalam Celcius: "))
fahrenheit = (celcius * 9 / 5) + 32
kelvin = (fahrenheit - 32) * 5 / 9 + 273.15

print(f"Suhu dalam Fahrenheit: {fahrenheit} ")
print(f"Suhu dalam Kelvin: {kelvin} K")
```

# 4. Buat program yang meminta input jarak dan kecepatan rata-rata, lalu hitung berapa lama waktu tempuh perjalanan tersebut.

image.png




```python
space = int(input("Masukkan jarak perjalanan (km): "))
velocity = int(input("Masukkan kecepatan rata-rata (km/jam): "))
time = space / velocity

print(f"Waktu tempuh perjalanan: {time} jam")
```

```python
secs = int(input("Masukkan waktu dalam format detik: "))
hrs = secs / 3600

print(f"{secs} detik sama dengan {hrs} jam")
```
