#1. Jelaskan apa itu algoritma iteration/loop!, dan juga apa saja syntax iteration/loop yang terdapat dalam python?

Simpelnya iteration/loop adalah perulangan, yang berarti algoritma ini berjalan secara berulang. Algoritma ini akan terus berjalan dan mengulang selama kondisinya bernilai `True`.

Algoritma loop memiliki 2 sintaks dalam python, yaitu:
- For
- While

Berbeda dengan bahasa lain yang memiliki 3 sintaks loop (do-while), di python perulangan hanya memiliki 2 sintaks loop.

#2. Jelaskan perbedaan antara syntax for dan while dalam python, beserta contoh kode sederhananya dari masing-masing syntax for dan while!

- For:
  > Simpelnya for adalah perulangan yang digunakan jika kita mengetahui batasan/jumlah pasti dari suatu perulangan.
- While:
  > While merupakan kebalikan dari for, jadi simpelnya while digunakan jika kita tidak mengetahui batasan atau kondisi pasti dari suatu perulangan.

```python
# Contoh for
for i in range(10):
  print(i)

# Contoh while
limit = int(input("Masukkan batas perulangan: "))
i = 0

while i <= limit:
  print(i)
  i += 1
```

#3. Jelaskan apa itu break, continue dan pass dalam python dan juga berikan contoh penggunaannya dalam iteration!

Dalam konteks iterasi/perulangan, fungsi dari `break`, `continue`, `pass` adalah sebagai berikut:
- break:
  > Pada perulangan break digunakan untuk menghentikan secara paksa perulangan, jadi operasi perulangan meski kondisinya masih terpenuhi akan secara paksa diberhentikan.
- continue:
  > Continue digunakan untuk melewati/skip suatu proses perulangan, jadi suatu langkah perulangan dapat dilewati dengan menggunakan sintaks ini.
- pass:
  > Pass berarti melewati. Sesuai artinya, pass berfungsi untuk melewati atau lebih tepatnya hanya lewat dan tidak melakukan apapun. Sintaks ini dapat digunakan untuk mengisi indentasi kosong agar tidak terjadi error.

```python
# Contoh break
for i in range(10):
  if i % 2 == 0:
    break

  print(i)

# Contoh continue
for i in range(10):
  if i % 2 != 0:
    continue

  print(i)

# Contoh pass
for i in range(10):
  pass
```
