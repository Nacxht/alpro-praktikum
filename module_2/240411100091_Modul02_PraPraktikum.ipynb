# 1. Apa itu Percabangan

Percabangan adalah suatu alur yang bercabang. Pemilihan tiap alur cabangnya ditentukan oleh suatu kondisi yang hanya memiliki 2 hasil, yaitu `true` dan `false`.

# 2. Pada saat apa kita perlu menggunakan percabangan

Percabangan akan berguna dan digunakan pada saat kita ingin menggunakan alur yang dapat ditentukan oleh suatu kondisi, contoh studi kasusnya adalah:
- Menentukan lulus atau tidaknya dengan nilai sebagai indikatornya.
- Menentukan layak/tidaknya seseorang membuat KTP.
- Menentukan layak/tidaknya seorang pembeli mendapatkan diskon.

# 3. Jelaskan syntax percabangan pada python

### **Sintaks percabangan yang sudah kuketahui ada 2, yaitu:**
- if - elif - else
- match

### **if - elif - else:**
sintaks ini dapat digunakan untuk membuat alur sesuai kondisi sebanyak 2 alur atau lebih, contoh penggunaan:
```py
nilai = int(input("Masukkan nilai anda: "))

if nilai >= 90:
  print("Nilai anda A")
elif nilai >= 80:
  print("Nilai anda B")
elif nilai >= 70:
  print("Nilai anda C")
else:
  print("Nilai anda D")
```
*Note: `elif` berarti `else-if`

### **Match:**
Match hampir sama seperti `if - elif - else`, bedanya match hanya mengecek kondisi melalui sama atau tidaknya nilai yang diberikan. Contohnya:
```py
menu = input("Masukkan menu anda dari 1 - 3")

match menu:
  case "1":
    print("anda memilih menu ke-1")
  case "2":
    print("anda memilih menu ke-2")
  case "3":
    print("anda memilih menu ke-3")
  case _:
    print("maaf pilihan anda tidak terdapat dalam menu")
```
Dari pengalaman saya, match memiliki performa yang lebih cepat dari sintaks `if`. Jadi saya lebih memiliki menggunakan `match` jika kondisi yang saya butuhkan merupakan kondisi yang hanya membutuhkan kecocokan nilai.
