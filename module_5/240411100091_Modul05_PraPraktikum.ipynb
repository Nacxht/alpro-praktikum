# 1. Diberikan string teks = "Belajar Python". Apa karakter yang terdapat pada indeks ke-8?

````python
# Variabel berisi string
stringTeks = "Belajar Python"

# Variabel berisi string yang telah diindeks
indexKe8 = stringTeks[8]

# Output
print(indexKe8)
````

Index ke-8 pada string teks "Belajar Python" adalah huruf "P".

Jadi semua index dalam pemrograman dimulai dari angka 0, jika melihat pada contoh diatas maka urutannya adalah seperti ini:
- "B" = 0
- "e" = 1
- "l" = 2
- "a" = 3
- "j" = 4
- "a" = 5
- "r" = 6
- " " = 7
- "P" = 8 <-
- "y" = 9
- "t" = 10
- "h" = 11
- "o" = 12
- "n" = 13

Jadi index ke-8 pada string teks tersebut adalah karakter "P", dan total panjang atau banyaknya karakter pada string teks tersebut adalah `13` karakter.

Jika di dalam python, kita bisa menggunakan metode indexing untuk mengetahui karakter apa dalam string ke-`n`.

Contoh kode:
```py
stringTeks = "Belajar Python"
indexKe8 = stringTeks[8]

print(indexKe8)
```

Penjelasan:
> tanda kurung siku-siku (brackets / [ ]) adalah suatu sintaks untuk melakukan indexing pada string dan list, dan format penggunaannya adalah `<string/list>[index]`.

# 2. Jika diberikan string kata = "Hello", bagaimana cara menggandakan string tersebut menjadi 5 kali dan tampilkan hasilnya?

````python
# Membuat variabel berisi string kata
stringKata = "Hello"

# Membuat variabel yang berisi string (yang telah digandakan)
stringGanda = stringKata * 5

# Output
print(stringGanda)
````

Jika diberikan string "Hello", dan anda ingin menggandakannya menjadi 5 kali atau berapapun itu? Caranya mudah, anda hanya perlu mengkalikan string tersebut dengan angka (sebanyak yang anda inginkan).

Mengikuti contoh yang diberikan, maka kode yang digunakan adalah sebagai berikut:
```py
stringKata = "Hello"
stringGanda = stringKata * 5 # Dikalikan 5

print(stringGanda)
```
akan menghasilkan output:
> `"HelloHelloHelloHelloHello"`

Penjelasan kode:
> Jadi variabel `stringKata` diatas memiliki tipe data `string`. Nah jadi string jika diberi proses aritmatik perkalian (*) akan berujung pada proses penggandaan string.

> Sebagai contoh adalah string `"Hello"` jika dikalikan dengan `5` akan menghasilkan string `Hello` sebanyak `5` kali.

> Apakah harus 5? Tidak, kita bisa menggunakan angka bernilai berapapun, dan hasil penggandaannya adalah sesuai dengan pengkali-nya.

# 3. Bagaimana cara menggabungkan dua string, nama_depan = "Rina" dan nama_belakang = "Sari", sehingga hasilnya menjadi "Rina Sari"?

````python
# Variabel string berisi nama depan & belakang
nama_depan = "Rina"
nama_belakang = "Sari"

# Variabel berisi string yang telah digabung
nama_lengkap1 = nama_depan + " " + nama_belakang        # Opsi 1
nama_lengkap2 = f"{nama_depan} {nama_belakang}"         # Opsi 2
nama_lengkap3 = " ".join([nama_depan, nama_belakang])   # Opsi 3

# Output
print(nama_lengkap1)
print(nama_lengkap2)
print(nama_lengkap3)
````

Untuk menggabungkan `2` string, kita bisa menggunakan banyak metode, tetapi yang paling cocok untuk menggabungkan `2` string adalah 3 metode berikut:
- String concatenation
- F-string
- List join



---
---

- Metode pertama (String concatenation):
  > Kita bisa menggunakan metode penggabungan string menggunakan dengan simbol aritmatik "+". Berikut adalah contoh penggunaannya:

  ```
  nama_depan = "Rina"
  nama_belakang = "Sari"

  nama_lengkap = nama_depan + " " + nama_belakang
  print(nama_lengkap)
  ```

  > Penjelasannya adalah variabel nama_depan berisi `"Rina"` dan variabel nama_belakang berisi `"Sari"` digabung menggunakan proses `+`, tetapi diantara penggabungan keduanya diberi lagi string `" "` (spasi) yang berfungsi memberikan spasi. Jika tidak menambahkan spasi, yang akan terjadi adalah string akan tergabung tanpa spasi (`RinaSari`).

- Metode kedua (F-string):
  > Format string juga dapat digunakan untuk menggabungkan kedua string, mengingat format string dapat memasukkan variabel secara bebas ke dalam penulisan string di python. Berikut adalha contoh penggunaannya:

  ```
  nama_depan = "Rina"
  nama_belakang = "Sari"

  nama_lengkap = f"{nama_depan} {nama_belakang}"
  print(nama_lengkap)
  ```
  > Penjelasannya adalah hampir sama dengan yang opsi ke-1, bedanya di metode ini kita tidak perlu menuliskan simbol `+` satu-persatu. Disini kita hanya menuliskan variabel `nama_depan`, lalu tambahkan spasi `(" ")`, dan yang terakhir tuliskan variabel `nama_belakang`. Dengan begini string sudah tergabung.

- Metode ketiga (List join):
  > Metode ketiga adalah metode yang bisa paling ribet tetapi fleksibel, mengapa fleksibel? Karena *seumpama* kita ingin menggabungkan string yang lebih dari 1 kita tidak perlu menambahkan perantara/penghubung antar string (contohnya: `","`, `" "`, dsb) jika menggunakan metode ini. Berikut contoh penggunaannya:

  ```
  nama_depan = "Rina"
  nama_belakang = "Sari"

  nama_lengkap = " ".join([nama_depan, nama_belakang])
  print(nama_lengkap)
  ```

  > Kode diatas harus menggunakan `list`.
    - Pertama kita tambahkan perantara/penghubung antar string yang kita inginkan, disini saya menggunakan string spasi `(" ")`.
    - Kedua gunakan `join()` yang merupakan built-in method dari python. Method ini berfungsi untuk menggabungkan `list`. Hasil akhirnya akan tampak seperti `" ".join()`.
    - Ketiga tambahkan string-string yang ingin digabung kedalam method `join()`. Disini saya memiliki variabel `nama_depan` dan `nama_belakang` yang ingin kumasukkan ke dalam method `join()`. Maka hasil akhirnya adalah `" ".join([nama_depan, nama_belakang])`.
    - Selamat! anda berhasil menggabungkan string menggunakan built-in method `join()`.

# 4. Diberikan string kalimat = "Pemrograman adalah seru!". Berapa jumlah karakter dalam string tersebut?

````python
# Variabel berisi string
stringKalimat = "Pemrograman adalah seru!"

# Wadah untuk menyimpan jumlah karakter
jumlahKarakter = 0

# Proses menghitung jumlah karakter
for i in stringKalimat:
  jumlahKarakter += 1

print(jumlahKarakter)
````

Jadi jumlah karakter pada string `"Pemrograman adalah seru!"` adalah `24` karakter:

- "a" = 17

Tetapi bagaimana panjang string yang kita miliki lebih tinggi lagi dari ini? Misal 100, 1000, bahkan 2000, apakah kita harus menghitungnya satu-persatu? Tidak! Jika kita belajar pemrograman, pasti ada cara untuk menyelesaikan permasalahan tersebut. Untuk menghitung jumlah karakter pada string pada string kita bisa menggunakan 2 metode:
- Menggunakan built-in function `len()`
- Menggunakan perulangan `for`


---
---


- Menggunakan built-in function `len()`:
  > Menggunakan metode ini adalah cara paling simpel, karena kita tinggal memasukkan string atau variabel berisi string ke dalam function tersebut. Contoh penggunaan:

  ```
  stringKalimat = "Pemrograman adalah seru!"
  jumlahKarakter = len(stringKalimat)

  print(jumlahKarakter)
  ```

  > Jadi jika kita menggunakan built-in function `len()`, fungsi ini akan secara otomatis menghitung panjang string kita, jumlah karakter pada string sama saja dengan panjang string.

- Menggunakan perulangan `for`:
  > Jika anda tidak percaya pada `len()` dan ingin metode yang lebih kompleks, anda bisa menggunakan perulangan `for` untuk menghitung jumlah karakter pada suatu string. Contoh penggunaan:

  ```
  stringKalimat = "Pemrograman adalah seru!"
  jumlahKarakter = 0

  for i in stringKalimat:
    jumlahKarakter += 1
  
  print(jumlahKarakter)
  ```

  > Jadi jika kita menuliskan `for i in <string>` pada perulangan loop, hal ini berarti variabel `i` akan berisi `tiap huruf` pada `tiap perulangan` (sesuai urutan perulangan). Seperti contohnya string "adi":
  - Putaran pertama `i` berisi `"a"`
  - Putaran kedua `i` berisi `"d"`
  - Putaran ketiga `i` berisi `"i"`
  Dengan begini jika kita isi `for` dengan `"Pemrograman adalah seru!"`, perulangan akan dijalankan sebanyak 24 kali (karena panjang stringnya adalah 24). Dengan begini tiap putaran kita dapat menambahkan variabel `jumlahKarakter` dengan angka 1, jadi ditiap putarannya variabel `jumlahKarakter` akan bertambah 1 nilai.
